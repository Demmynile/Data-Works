<h3>Reporting using Plotly Dash/Google Data studio</h3>

```python
import dash
from dash.dependencies import Input,Output
import dash_core_components as dcc
import dash_html_components as html
import pandas as pd
import plotly.express as px
from matplotlib import pyplot as plt




#include the css
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']



#intializing the dash library
app = dash.Dash(__name__ , external_stylesheets = external_stylesheets)

#prepare the data for Reporting with reactive functions inclusive
data = pd.read_csv('data_cleaned.csv')
fig = px.bar(data , x = 'PRODUCT CATEGORY' , y = 'UNIT PRICE')
fig2 = px.scatter(data , x = 'UNIT PRICE' , y = 'DISCOUNT')

#creating the layout components using html 
app.layout =  html.Div(children = [
    html.H2(children = 'Ecommerce Report' , style = {
       'textAlign': 'center'

       

    }
    ),
    
    html.H6(children = 'Here are the basic sales report generated from last year to this year',
       style  = {
           'textAlign' :'center'
       }
    
    ),
    dcc.Graph(

        figure = fig
    ),

    html.H6(children = "The Below report shows the relationship between Price and Discount ", 
    style = {
        'textAlign':'center'

     }
    
    ),
    dcc.Graph(

      figure = fig2


    ),

])
#Running my application
if __name__ == '__main__':
    app.run_server(debug = True)
```
